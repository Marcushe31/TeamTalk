```python
const express = require('express');
const app = express();
const PORT = process.env.PORT || 3000;

// Middleware to parse JSON requests
app.use(express.json());

// Example API endpoint to get all cards
app.get('/cards', (req, res) => {
    // For simplicity, we'll use a static array, but you would fetch this data from your database
    const cards = [
        { id: 1, name: 'Jack Schneider', position: 'Quarter Back' },
        { id: 2, name: 'Ty Olsen', position: 'Wide Receiver' },
        // ...other cards
    ];
    res.json(cards);
});

// Start the server
app.listen(PORT, () => {
    console.log(`Server is running on http://127.0.0.1:4200/student/#`);
});

const express = require('express');
const app = express();
const PORT = process.env.PORT || 3000;

app.use(express.json());

// 使用数组来临时存储卡片数据
let cards = [
    { id: 1, name: 'Jack Schneider', position: 'Quarter Back' },
    { id: 2, name: 'Ty Olsen', position: 'Wide Receiver' },
    // ...other cards
];

// 获取所有卡片
app.get('/cards', (req, res) => {
    res.json(cards);
});

// 添加新卡片
app.post('/cards', (req, res) => {
    const newCard = req.body;
    newCard.id = cards.length + 1;  // 为新卡片分配一个 ID
    cards.push(newCard);
    res.json(newCard);
});

// 删除卡片
app.delete('/cards/:id', (req, res) => {
    const id = parseInt(req.params.id, 10);
    cards = cards.filter(card => card.id !== id);
    res.json({ message: 'Card deleted successfully!' });
});

app.listen(PORT, () => {
    console.log(`Server is running on http://127.0.0.1:4200/student/#`);
});
```
